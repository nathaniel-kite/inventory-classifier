```python
import pandas
import csv
import numpy as np
import random

df = pandas.read_csv("phoible-2.0/data/phoible.csv")

df
```

```python
# Get all phonemes
phonemes = df['Phoneme'].unique()

print(phonemes)
print(len(phonemes))
```

```python
korean = df[df["ISO6393"] == "kor"]

korean_inventories = korean['InventoryID'].unique()

korean_inventories
```

```python
glottocodes = df['Glottocode'].unique()

glottocodes
```

```python
languoids = []

for code in glottocodes:
    inventories = df[df['Glottocode'] == code]['InventoryID'].unique()
    languoids.append({'glottocode': code, 'inventories': inventories})

selected_inventories = []

for lang in languoids:
    # For some reason there is a 'NaN' glottocode in here, with no inventories associated with it
    if len(lang['inventories']) != 0:
        selected_inventories.append(random.choice(lang['inventories']))
        
selected_inventories
```

```python
df_new = df[df['InventoryID'].isin(selected_inventories)]

df_new
```
